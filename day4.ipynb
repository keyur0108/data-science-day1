```python
import pandas as pd
```

```python
df=pd.read_csv("QueryResults.csv",header=0,names=["DATE","TAG","POSTS"])
```

```python
df
```

```python
df.head()
```

```python
df.tail()
```

```python
df.count()
```

```python
df.groupby(["TAG"]).sum()
```

```python
df.groupby(["TAG"]).count()
```

```python
df.DATE[1]
```

```python
type(df.DATE[1])
```

```python
df.DATE=pd.to_datetime(df.DATE)
```

```python
type(df.DATE[1])
```

```python
df.head()
```

```python
reshaped_df = df.pivot(index="DATE",columns="TAG",values="POSTS")
```

```python
reshaped_df
```

```python
reshaped_df.count()
```

```python
reshaped_df.fillna(0,inplace=True)
```

```python
reshaped_df.head()
```

```python
reshaped_df.isna()
```

```python
reshaped_df.isna().sum()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(reshaped_df.index,reshaped_df["java"])
```

```python
plt.figure(figsize=(16,10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.plot(reshaped_df.index,reshaped_df["java"])
```

```python
plt.figure(figsize=(16,10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel("Date",fontsize=14)
plt.ylabel("No of Posts",fontsize=14)
plt.ylim(0,35000)
plt.plot(reshaped_df.index,reshaped_df["java"])
```

```python
plt.figure(figsize=(16,10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel("Date",fontsize=14)
plt.ylabel("No of Posts",fontsize=14)
plt.ylim(0,35000)
plt.plot(reshaped_df.index,reshaped_df["java"])
plt.plot(reshaped_df.index,reshaped_df["python"])
plt.plot(reshaped_df.index,reshaped_df["c"])
plt.plot(reshaped_df.index,reshaped_df["c++"])
plt.plot(reshaped_df.index,reshaped_df["php"])
plt.plot(reshaped_df.index,reshaped_df["ruby"])
plt.plot(reshaped_df.index,reshaped_df["swift"])
```

```python
plt.figure(figsize=(16,10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel("Date",fontsize=14)
plt.ylabel("No of Posts",fontsize=14)
plt.ylim(0,35000)
for column in reshaped_df.columns:
    plt.plot(reshaped_df.index,reshaped_df[column],label=reshaped_df[column],linewidth=3)
plt.legend(fontsize=16)
plt.show(block=False)
```

```python
plt.figure(figsize=(16,10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel("Date",fontsize=14)
plt.ylabel("No of Posts",fontsize=14)
plt.ylim(0,35000)
for column in reshaped_df.columns:
    plt.plot(reshaped_df.index,reshaped_df[column],label=column,linewidth=3)
plt.legend(fontsize=16)
```

```python
rolling_df=reshaped_df.rolling(window=6).mean()
plt.figure(figsize=(16,10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel("Date",fontsize=14)
plt.ylabel("No of Posts",fontsize=14)
plt.ylim(0,35000)
for column in rolling_df.columns:
    plt.plot(rolling_df.index,rolling_df[column],label=column,linewidth=3)
plt.legend(fontsize=16)
```
